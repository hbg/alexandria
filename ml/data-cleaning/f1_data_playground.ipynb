```python
import pandas as pd
import kagglehub
import os
```

```python
path = kagglehub.dataset_download("rohanrao/formula-1-world-championship-1950-2020")
```

```python
os.listdir(path)
```

```python
df = pd.read_csv(os.path.join(path, "results.csv"))
df.head()
```

```python
df_drivers = pd.read_csv(os.path.join(path, "drivers.csv"))
df_drivers.head()
```

```python
df_constructors = pd.read_csv(os.path.join(path, "constructors.csv"))
df_constructors.head()
```

```python
df_seasons = pd.read_csv(os.path.join(path, "seasons.csv"))
df_seasons.head()
```

```python
df_seasons[df_seasons.year > 2010].head()
```

```python
df_races = pd.read_csv(os.path.join(path, "races.csv"))
df_races.head()
```

```python
df = df.merge(df_drivers, on="driverId", how="inner")
df.head()
```

```python
df_races[df_races.year > 2010]
```

```python
new_races = df[df.raceId.isin(df_races[df_races.year > 2010].raceId)]
```

```python
print(f"erased: {(len(df) - len(new_races))/len(df)*100:.2f}% of the data")
```

```python
df_winners = new_races[new_races.positionOrder == 1]
df_winners.head()
```

```python
df_winner_agg = df_winners.groupby("code").size().sort_values(ascending=False)
df_winner_agg.head(50)
```

```python
updated = pd.merge(df_winner_agg.rename('total_wins'), df, on='code', how='inner')
updated.head(100)
```

```python
from matplotlib import pyplot as plt
```

```python
plt.scatter(updated.positionOrder, updated.points)
```

```python
updated['fullName'] = updated.forename + " " + updated.surname
updated.head()
```

```python
df_circuits = pd.read_csv(os.path.join(path, "circuits.csv"))
df_circuits.head()
```

```python
updated = pd.merge(updated, df_races, on='raceId', how='inner')
```

```python
updated = pd.merge(updated, df_circuits, on='circuitId', how='inner')
```

```python
updated.head()
```

```python
updated.groupby('circuitId').size().sort_values(ascending=False).head(10)
```

```python
updated_circuit_9 = updated[updated.circuitId == 9]
updated_circuit_9 = updated_circuit_9[updated_circuit_9.position != '\\N']
updated_circuit_9 = updated_circuit_9[updated_circuit_9.fastestLapTime != '\\N']

updated_circuit_9['position'] = updated_circuit_9['position'].astype(int)
```

```python
# reformat fastest lap time to seconds from format mm:ss.sss
updated_circuit_9['fastestLapTimeSeconds'] = updated_circuit_9['fastestLapTime'].str.split(':').apply(
    lambda x: int(x[0]) * 60 + float(x[1])
)
updated_circuit_9['fastestLapTimeSeconds'] = updated_circuit_9['fastestLapTimeSeconds'].astype(float)
```

```python
ax = updated_circuit_9.plot.scatter(x='position', y='fastestLapTimeSeconds', c='red', alpha=0.5)
for x, y, racer in zip(updated_circuit_9.position, updated_circuit_9.fastestLapTimeSeconds, updated_circuit_9.code):
    ax.text(x, y, f'{racer}')
```

```python
updated_circuit_9_recent = updated_circuit_9[updated_circuit_9.year > 2010]
ax = updated_circuit_9_recent.plot.scatter(x='position', y='fastestLapTimeSeconds', c='red', alpha=0.5)
for x, y, racer in zip(updated_circuit_9_recent.position, updated_circuit_9_recent.fastestLapTimeSeconds, updated_circuit_9.code):
    ax.text(x, y, f'{racer}')
```

```python
updated_circuit_9_recent = updated_circuit_9[updated_circuit_9.year > 2020]
ax = updated_circuit_9_recent.plot.scatter(x='position', y='fastestLapTimeSeconds', c='red', alpha=0.5)
for x, y, racer in zip(updated_circuit_9_recent.position, updated_circuit_9_recent.fastestLapTimeSeconds, updated_circuit_9_recent.code):
    ax.text(x, y, f'{racer}')
```

```python
updated_circuit_9_recent = updated_circuit_9_recent[updated_circuit_9_recent.total_wins > 50]
ax = updated_circuit_9_recent.plot.scatter(x='position', y='fastestLapTimeSeconds', c='red', alpha=0.5)
for x, y, racer in zip(updated_circuit_9_recent.position, updated_circuit_9_recent.fastestLapTimeSeconds, updated_circuit_9_recent.code):
    ax.text(x, y, f'{racer}')
```

```python
wins_by_year = updated[updated.positionOrder == 1].groupby(['year', 'code']).size().to_frame(name='wins').reset_index()
wins_by_year
```

```python
ham_wins = wins_by_year[wins_by_year.code == 'HAM']
ham_wins
```

```python
avg_position_by_year = updated.groupby(['year', 'code']).agg({'positionOrder': 'mean'}).reset_index().rename(
    columns={'positionOrder': 'avg_position'})
avg_position_by_year
```

```python
ham_avg_position_by_year = avg_position_by_year[avg_position_by_year.code == 'HAM']
ham_avg_position_by_year
```

```python
# If ham_wins is empty, it means there are no wins for HAM in the dataset for that year
merged_ham_data = pd.merge(ham_wins, ham_avg_position_by_year, on=['year', 'code'], how='outer')
merged_ham_data = merged_ham_data.fillna(0)
merged_ham_data = merged_ham_data.sort_values(by='year')
merged_ham_data.head()
```

```python
ax = merged_ham_data.plot.scatter(x='avg_position', y='wins', color='blue', alpha=0.5, title = 'Lewis Hamilton Wins vs Average Position')
for x, y, year in zip(merged_ham_data.avg_position, merged_ham_data.wins, merged_ham_data.year):
    ax.text(x, y, f'{year}', fontsize=8, ha='center', va='bottom')
```
